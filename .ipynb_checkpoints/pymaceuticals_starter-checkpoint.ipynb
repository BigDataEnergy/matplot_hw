## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
from scipy.stats import linregress
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
df = pd.merge(mouse_metadata, study_results, how = 'outer', on = 'Mouse ID')

# Display the data table for preview
df
```

```python
# Checking the number of mice.
count_df = df.drop_duplicates('Mouse ID', keep= 'first', inplace= False)['Mouse ID'].count()
count_df
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
group_df = df.groupby('Mouse ID').count()
group_df = group_df.sort_values('Timepoint', ascending= False)    
group_df
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicate_df = group_df[group_df.Timepoint > 10].filter(['Mouse ID'])
duplicate_df = pd.merge(df, duplicate_df, on = 'Mouse ID')
duplicate_df
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
df = df[~df['Mouse ID'].isin(duplicate_df['Mouse ID'])]
df
```

```python
# Checking the number of mice in the clean DataFrame.
count_df = df.drop_duplicates('Mouse ID', keep= 'first', inplace= False)['Mouse ID'].count()
count_df
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
regimen_df = df.groupby('Drug Regimen')

reg_mean = regimen_df['Tumor Volume (mm3)'].mean()
reg_med = regimen_df['Tumor Volume (mm3)'].median()
reg_var = regimen_df['Tumor Volume (mm3)'].var()
reg_std = regimen_df['Tumor Volume (mm3)'].std()
reg_sem = regimen_df['Tumor Volume (mm3)'].sem()

summary_df = pd.DataFrame({'Mean': reg_mean, 'Median': reg_med, 'Variance': reg_var, 'Standard Deviation': reg_std, 'SEM': reg_sem})
summary_df
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
agg_df = df.groupby('Drug Regimen')
# Using the aggregation method, produce the same summary statistics in a single line
agg_df.agg(['mean', 'median','var', 'std', 'sem'])['Tumor Volume (mm3)']
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
drugs = ['Capomulin','Ramicane','Infubinol','Ceftamin']
# Start by getting the last (greatest) timepoint for each mouse
max_tp = pd.DataFrame(df.groupby('Mouse ID')['Timepoint'].max()).rename(columns={'Timepoint': 'Max Timepoint'})

# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
merged_df = pd.merge(df, max_tp, on = 'Mouse ID')
merged_df.loc[merged_df['Drug Regimen'].isin(drugs)]
```

```python
# Put treatments into a list for for loop (and later for plot labels)

### Already created in previous cell

# Create empty list to fill with tumor vol data (for plotting)
tumor_val = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 

for drug in drugs:  

    # Locate the rows which contain mice on each drug and get the tumor volumes
    temp_df = merged_df.loc[merged_df['Drug Regimen'] == drug]
    final_vol_df = temp_df.loc[temp_df['Timepoint'] == temp_df['Max Timepoint']]
    
    values = final_vol_df['Tumor Volume (mm3)']
    tumor_val.append(values)
    
    # add subset 
    
    
    
    ### Calc IQR for each drug
    quartiles = values.quantile([.25, .5, .75])
    lowerq = quartiles[.25]
    upperq = quartiles[.75]
    iqr = upperq-lowerq
    print(f'IQR for {drug}: {iqr}')
    
    # Determine outliers using upper and lower bounds
    lowerb = lowerq - (1.5*iqr)
    upperb = upperq + (1.5*iqr)
    outlier_count = (values.loc[(final_vol_df['Tumor Volume (mm3)']>= upperb) | (final_vol_df['Tumor Volume (mm3)']<= lowerb)]).count()
    
    print(f'Number of outliers for {drug}: {outlier_count}')
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest

plt.boxplot(tumor_val)
plt.xticks([1,2,3,4],['Capomulin','Ramicane', 'Infubinol', 'Ceftamin'])
plt.title('Final Tumor Volumes')
plt.ylabel('Final Tumor Volume (mm3)')
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a mouse treated with Capomulin

#select random mouse under Capomulin treatment
mouse_df = df.loc[df['Drug Regimen'] == 'Capomulin'].sample()
mouse_df = mouse_df.reset_index(drop=True)


#single out mouse
ID = mouse_df.iloc[0,0]
mouse_data = df.loc[df['Mouse ID'] == ID]

#create plot
plt.plot(mouse_data['Timepoint'], mouse_data['Tumor Volume (mm3)'], marker = 'o')
plt.xlabel('Time (days)')
plt.ylabel('Tumor Volume (mm3)')
plt.title(f'Capomulin Treatment for mouse' + ' ' + ID)
plt.show()
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen
capo_df = df.loc[df['Drug Regimen'] == 'Capomulin']

avg_vol_df = pd.DataFrame(capo_df.groupby('Mouse ID')['Tumor Volume (mm3)'].mean()).reset_index().rename(columns={'Tumor Volume (mm3)': 'Average Tumor Volume'})
merged_df = pd.merge(capo_df, avg_vol_df, on = 'Mouse ID')
avg_vol_df = merged_df[['Weight (g)', 'Average Tumor Volume']].drop_duplicates()

#define variables/create plot
x = avg_vol_df['Weight (g)']
y = avg_vol_df['Average Tumor Volume']

plt.scatter(x,y)
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.title('Average Tumor Volume by Weight')
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
coeff = st.pearsonr(x,y)

#linear regression
(m, b, r, p, std) = linregress(x, y)
reg_values = m*x + b
line = "y = " + str(round(m,2)) + "x + " + str(round(b,2))

# generate plot
plt.scatter(x,y)
plt.plot(x,reg_values,"r-")


#labels and annotation
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.title('Average Tumor Volume by Weight')
plt.annotate(line,(20,35),fontsize=12,color="red")
plt.show()
```
