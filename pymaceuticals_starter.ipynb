## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
df = pd.merge(mouse_metadata, study_results, how = 'outer', on = 'Mouse ID')

# Display the data table for preview
df
```

```python
# Checking the number of mice.
count_df = df.drop_duplicates('Mouse ID', keep= 'first', inplace= False)['Mouse ID'].count()
count_df
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
group_df = df.groupby('Mouse ID').count()
group_df = group_df.sort_values('Timepoint', ascending= False)    
group_df
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicate_df = group_df[group_df.Timepoint > 10].filter(['Mouse ID'])
duplicate_df = pd.merge(df, duplicate_df, on = 'Mouse ID')
duplicate_df
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
df = df[~df['Mouse ID'].isin(duplicate_df['Mouse ID'])]
df
```

```python
# Checking the number of mice in the clean DataFrame.
count_df = df.drop_duplicates('Mouse ID', keep= 'first', inplace= False)['Mouse ID'].count()
count_df
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
regimen_df = df.groupby('Drug Regimen')

reg_mean = regimen_df['Tumor Volume (mm3)'].mean()
reg_med = regimen_df['Tumor Volume (mm3)'].median()
reg_var = regimen_df['Tumor Volume (mm3)'].var()
reg_std = regimen_df['Tumor Volume (mm3)'].std()
reg_sem = regimen_df['Tumor Volume (mm3)'].sem()

summary_df = pd.DataFrame({'Mean': reg_mean, 'Median': reg_med, 'Variance': reg_var, 'Standard Deviation': reg_std, 'SEM': reg_sem})
summary_df
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
agg_df = df.groupby('Drug Regimen')
# Using the aggregation method, produce the same summary statistics in a single line
agg_df.agg(['mean', 'median','var', 'std', 'sem'])['Tumor Volume (mm3)']
```
